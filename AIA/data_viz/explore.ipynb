````python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

records = []
for _, r in df.iterrows():
    header = f"{r.Id}|{r.Fungi_species}|{r.Isolation_Source}|{r.Country}"
    rec = SeqRecord(Seq(r.Sequence), id=header, description="")
    records.append(rec)

SeqIO.write(records, "fungi.fna", "fasta")
````

````python
from Bio import AlignIO
aln = AlignIO.read("fungi_aligned.fna", "fasta")
seqs   = [str(r.seq) for r in aln]
labels = [r.id for r in aln]
````

````python
import numpy as np

alpha = "ACGT-"                                    # 5 символов (выр. с gap’ом)
code  = {b:i for i,b in enumerate(alpha)}

M = np.array([[code.get(ch.upper(), 4)
               for ch in s] for s in seqs], dtype=np.uint8)

onehot = (np.arange(5) == M[...,None]).astype("float32")
X = onehot.reshape(len(seqs), -1)
````

````python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

# coords = PCA(n_components=3, svd_solver="randomized").fit_transform(X)

# coords = TSNE(n_components=3, metric="cosine", perplexity=10,
#               learning_rate='auto', init='pca', random_state=42).fit_transform(X)


import umap

umap3d = umap.UMAP(
    n_components=3,
    n_neighbors=15,
    min_dist=0.1,
    metric="cosine",
)
coords = umap3d.fit_transform(X)
````

````python
coords_df = pd.DataFrame(coords, columns=["x3d", "y3d", "z3d"])
df = pd.concat([df, coords_df], axis=1)

# вариант 2 ─ напрямую присвоить
# df[["x3d", "y3d", "z3d"]] = coords

# сохраняем
df.to_csv("fungi_with_coords.csv", index=False)
````
