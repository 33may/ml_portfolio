```python
dataset = []
```

```python
import time

from AIA.rl.lander.envs import PushTEnv
import numpy as np

env = PushTEnv(obs_type="state", render_mode="human")
teleop = env.teleop_agent()

obs, info = env.reset()
done = False

prev_reward = 0
total_reward = 0

while True:
    time.sleep(0.2)
    action = teleop.act(obs)

    if action is not None:
        new_state, reward, terminated, truncated, info = env.step(action)

        prev_t_pos = obs[2:3]
        new_t_pos = new_state[2:3]

        d_move = np.sqrt(np.sum((prev_t_pos - new_t_pos)**2))

        d_move_reward = d_move / 1000

        buffer_reward = reward

        reward -= prev_reward
        reward += d_move_reward

        prev_reward = buffer_reward

        done = terminated or truncated
        dataset.append((obs, action, reward, done, new_state))
        obs = new_state
        total_reward += reward

        print(reward)

        if done:
            print(f"Episode finished total reward: {total_reward}.")
            total_reward = 0
            prev_reward = 0
            obs, info = env.reset()
            done = False

    env.render()
```

```python
env.close()
```

```python
import pickle

with open("demonstrations.pkl", "wb") as f:
    pickle.dump(dataset, f, protocol=pickle.HIGHEST_PROTOCOL)
```
